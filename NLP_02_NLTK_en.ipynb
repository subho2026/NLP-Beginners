**Task**

Find occurences and context for the following list of texts and words



*   **text** `Sense and Sensibility` - **word** `affection`
*   **text** `book of Genesis` - **word** `lived`
*   **text** `Inaugural Address Corpus` - **words** `nation`, `terror`, `god`
*   **text** `NPS Chat Corpus` - **word** `im`, `ur`, `lol`

```python
#write your code here

import nltk
nltk.download('book')
from nltk.book import *

def find_occurrences_and_context(text_corpus, target_words, context_window=15):
    for text in text_corpus:
        for word in target_words:
            concordance = text.concordance_list(word, width=context_window)
            if len(concordance) > 0:
                print(f"Text: {text.name}")
                print(f"Word: {word}")
                print("Context:")
                for entry in concordance:
                    print(entry.line)
                print("\n")

# Define the list of texts and target words
text_corpus_list = [text1, text4, text5, text7]
target_words_list = ["affection", "lived", "nation", "terror", "god", "im", "ur", "lol"]

# Find occurrences and context
find_occurrences_and_context(text_corpus_list, target_words_list)

```

Some extra conditions



* `s.startswith(t)`	test if `s` starts with `t`

* `s.endswith(t)`	test if `s` ends with `t`

* `t in s`	test if `t` is a substring of `s`

* `s.islower()`	test if `s` contains cased characters and all are lowercase

* `s.isupper()`	test if `s` contains cased characters and all are uppercase

* `s.isalpha()`	test if `s` is non-empty and all characters in `s` are alphabetic

* `s.isalnum()`	test if `s` is non-empty and all characters in `s` are alphanumeric

* `s.isdigit()`	test if `s` is non-empty and all characters in `s` are digits

* `s.istitle()`	test if `s` contains cased characters and is titlecased (i.e. all words in `s` have initial capitals)

**6)** Find all the four-letter words in the Chat Corpus (text5).

With the help of a frequency distribution (FreqDist), show these words in decreasing order of frequency.

```python
# write your code here
import nltk

# Load the text
from nltk.book import text5

# Find all four-letter words
four_letter_words = [word.lower() for word in text5 if len(word) == 4 and word.isalpha()]

# Create a frequency distribution for four-letter words
fdist = nltk.FreqDist(four_letter_words)

# Display four-letter words in decreasing order of frequency
sorted_words = sorted(fdist.items(), key=lambda item: item[1], reverse=True)
for word, frequency in sorted_words:
    print(word, frequency)
```

**8)** Write expressions for finding all words in text6 that meet the conditions listed below.

*  Ending in `ise`
* Containing the letter `z`
* Containing the sequence of letters `pt`
* Having all lowercase letters except for an initial capital (i.e., `titlecase`)

```python
# write your code here
import re

# Find words ending in "ise"
words_ending_in_ise = [word for word in text6 if re.search(r'ise$', word)]
# Find words containing the letter "z"
words_containing_z = [word for word in text6 if re.search(r'z', word)]
# Find words containing the sequence "pt"
words_containing_pt = [word for word in text6 if re.search(r'pt', word)]
# Find words in titlecase
titlecase_words = [word for word in text6 if word.istitle()]

print(words_ending_in_ise)
print(words_containing_z)
print(words_containing_pt)
print(titlecase_words)
```
